```python
import torch
from torchvision import transforms
from PIL import Image
import os
```

```python
def color_jitter(img_path,save_path):
    # 画像を読み込む
    img = Image.open(img_path)

    # コントラストをランダムに変更するトランスフォームを定義
    color_jitter = transforms.ColorJitter(brightness=0.5, contrast=(0.3, 2.0), saturation=0.5, hue=0)

    # 画像に変換を適用
    transformed_img = color_jitter(img)

    # 変換後の画像を表示または保存
    #transformed_img.show()  # 画像を表示
    print(save_path)
    transformed_img.save(save_path)  # 画像を保存
```

```python
data_path = "ultralytics/Image_label/val/images"
data_dir = os.listdir(data_path)
```

```python
for path in data_dir:
    img_path = os.path.join(data_path,path)
    
    save_path = os.path.join(data_path,"cJ_" + path)
    color_jitter(img_path,save_path)
```

```python
label = "ultralytics/Image_label/val/labels"
label_dir = os.listdir(label)
```

```python
# 必要なモジュールをインポート
import os

# ファイルの読み込み
def save_label(label_path,i):
    with open(label_path, "r") as file:
        file_contents = file.read()

    # 新しいファイル名を指定
    new_filename = "cJ_"+ i

    
    new_path = os.path.join(label, new_filename)

    # ファイルの書き込み
    with open(new_path, "w") as new_file:
        new_file.write(file_contents)

    print(f"ファイル {new_filename} が作成されました。")
```

```python
for i in label_dir:
    label_path = os.path.join(label,i)
    print(label_path)
    save_label(label_path,i)
```
