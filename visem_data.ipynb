```python
for num in data_index:
    data_list = os.path.join(data_dir,num)
    root_dir = os.path.join(data_list+'/labels')
    for i,path in enumerate(os.listdir(data_list+'/labels')):
        print(i,path)
        if i % 30 == 1:
            f = open(os.path.join(root_dir,path),'r')
            data = f.read()
            print(data)
            
            with open(os.path.join(save_dir,path),'w') as f:
                f.write(data)
        
            
            
```
